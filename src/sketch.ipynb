```julia
using MyCalculus
using Plots
using Symbolics
```

## Appply Euler's method for the following First-Order Differential Equation:




$fun(t, y) = y^2 -t$



```julia
fun(t, y) = y^2 -t
```

```julia
x, y = EulerMethod(fun, -1.0, -0.5, 0.5, 5)
```

```julia
plot(x,y, ylim=(-0.5,0.5), xlim=(-1.0,1.0), framestyle = :origin)
```

### We can also use Symbolics by passing Symbolics variables to the function:


```julia
@variables x, y, step
t, y =  EulerMethod(fun, x, y, step, 5)
```

```julia
y
```

## Now we reduce step size:

```julia
x1, y1 = EulerMethod(fun, -1.0, -0.5, 0.5, 25)
x2, y2 = EulerMethod(fun, -1.0, -0.5, 0.25, 100)
x3, y3 = EulerMethod(fun, -1.0, -0.5, 0.125, 1000)
x4, y4 = EulerMethod(fun, -1.0, -0.5, 0.0125, 1000)
```

```julia
plot(x1,y1, ylim=(-4.0,0.5), xlim=(-1.0,10.0), framestyle = :origin) # you can see some instability around x=8.0
plot!(x2,y2, ylim=(-4.0,0.5), xlim=(-1.0,10.0), framestyle = :origin) 
plot!(x3,y3, ylim=(-4.0,0.5), xlim=(-1.0,10.0), framestyle = :origin)
plot!(x4,y4, ylim=(-4.0,0.5), xlim=(-1.0,10.0), framestyle = :origin)
```
